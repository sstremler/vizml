# KernelPerceptron

```python
import numpy as np
from IPython.display import IFrame, HTML
from vizml.perceptron import KernelPerceptron
```

```python
np.random.seed(0)

N = 200
X = np.random.normal(loc=0, scale=1, size=(N,2))

# generate classes
# let the class be -1 under the mirror of the hyperbolic cosine function
# and +1 above
y = np.where(X[:,1] <= -1 * np.cosh(X[:,0]) + 1.5, -1, 1)
X[y == 1,1] += 0.5
```

### Gaussian kernel

```python
clfG = KernelPerceptron(kernel="gaussian", kernel_param=1)
alpha = clfG.fit(X, y)
```

```python
animG = clfG.plot_animation(interval=500, notebook=True, smoothness=400, xinch=5.5, yinch=5)
HTML(animG.to_html5_video())
```

### Polynomial kernel

```python
clfP = KernelPerceptron(kernel="polynomial", kernel_param=2)
alpha = clfP.fit(X, y)
```

```python
animP = clfP.plot_animation(interval=500, notebook=True, smoothness=400, xinch=5.5, yinch=5)
HTML(animP.to_html5_video())
```
