# VotedPerceptron

```python
import numpy as np
from vizml.perceptron import VotedPerceptron
from IPython.display import IFrame, HTML
np.random.seed(0)
```

## Test 1

```python
N = 100
X = np.random.normal(loc=1.2, scale=1, size=(N,2))
X = np.row_stack((X, np.random.normal(loc=0, scale=1, size=(N,2))))
y = np.row_stack((np.full((N,1), -1), np.full((N,1), 1))).ravel()

clf = VotedPerceptron(eta=0.5, epoch=10)
clf.fit(X, y)

anim = clf.plot_animation(interval=400, notebook=True, smoothness=400)
HTML(anim.to_html5_video())
```

```python
print("Training score: ", str(clf.score() * 100), "%")
```

## Test 2

```python
N = 200
X = np.random.normal(loc=0, scale=1, size=(N,2))

# generate classes
# let the class be -1 under the mirror of the hyperbolic cosine function
# and +1 above
y = np.where(X[:,1] <= -1 * np.cosh(X[:,0]) + 1.5, -1, 1)
X[y == 1,1] += 0.5

clf = VotedPerceptron(eta=0.5, epoch=1)
clf.fit(X, y)

anim = clf.plot_animation(interval=400, notebook=True, smoothness=400)
HTML(anim.to_html5_video())
```

```python
print("Training score: ", str(clf.score() * 100), "%")
```
