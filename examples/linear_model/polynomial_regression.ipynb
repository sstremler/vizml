# PolynomialRegression

```python
import numpy as np
from vizml.linear_model import PolynomialRegression
from IPython.display import IFrame, HTML
```

```python
np.random.seed(0)

X = np.linspace(-4, 2.5, 100).reshape(100,1)
y = X**3/4 + 3*X**2/4 - 3*X/2 + np.random.normal(loc=0, scale=0.2, size=(100,1))
y = y.ravel()
```

```python
clf = PolynomialRegression(eta=0.001, maxit=500, alpha=0, degree=3)
w = clf.fit(X, y)
```

## Plot

```python
anim = clf.plot_animation(interval=20, notebook=True, xinch=5.5, yinch=5)
HTML(anim.to_html5_video())
```
